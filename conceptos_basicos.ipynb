**CONCEPTOS BÁSICOS DE PYTHON**

El siguiente proyecto describe los ejercicios para los conceptos básicos de programación en PYTHON



```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
np.random.seed(42) # for reproducibility
marketing_investment = np.random.rand(100) * 1000
sales = marketing_investment * 2 + np.random.randn(100) * 200 + 500

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(marketing_investment, sales, color='red')
plt.title('Sales vs. Marketing Investment')
plt.xlabel('Marketing Investment')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

**BUCLES O CICLOS**

```python
# Ejemplo Bucle while

numero = 1

while numero <= 5:
    print(numero)
    numero += 1
```

```python
# Ejemplo Bucle for
frutas = ["manzana", "banana", "cereza"]

for ejemplo in frutas:
    print("A mi me justa la", ejemplo)
```

```python
# Ejemplo for variable numerica

numeros = [1, 2, 3, 4, 5]

for resultado in numeros:
        print(resultado*10)
```
