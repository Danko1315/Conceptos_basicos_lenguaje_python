```python
# cargar la libreria pandas para importar la base de datos .csv .xlsx
import pandas as pd
# crear una variable que se llame "datos" y a eso le vamos a cargar la base de datos
datos = pd.read_csv("/content/drive/MyDrive/unidad 2/datos/Reporte_Delito_Violencia_Intrafamiliar_Polic_a_Nacional.csv")
datos.head()
```

```python
datos.info()
```

```python
print("Unique values for ARMAS MEDIOS:")
print(datos['ARMAS MEDIOS'].unique())
print("\nUnique values for GENERO:")
print(datos['GENERO'].unique())
print("\nUnique values for GRUPO ETARIO:")
print(datos['GRUPO ETARIO'].unique())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick

# Group by GENERO and sum CANTIDAD
gender_counts = datos.groupby('GENERO')['CANTIDAD'].sum().reset_index()

# Create the bar chart
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='GENERO', y='CANTIDAD', data=gender_counts, color='green')
plt.title('Cantidad por Género')
plt.xlabel('Género', fontweight='bold')
plt.ylabel('Cantidad', fontweight='bold')

# Format y-axis labels with comma separator
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

plt.show()
```
